```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ipl = pd.read_csv('matches.csv')
ipl.head()
```

```python
#number of rows and columns
ipl.shape
```

```python
#Top 10 player of match
pom = pd.DataFrame(ipl['player_of_match'].value_counts()[0:10]).reset_index()
pom
```

```python
plt.figure(figsize=(15,5))
plt.bar(pom['index'] , pom['player_of_match'],color = 'b')
plt.title("Count of top 10 player of match")
plt.xlabel("Player Name")
plt.show()
```

```python
#Frequency of result
pd.DataFrame(ipl['result'].value_counts()).reset_index()
```

```python
#Toss winner
tw = pd.DataFrame(ipl['toss_winner'].value_counts()).reset_index()
tw
```

```python
plt.figure(figsize=(10,5))
plt.barh(tw['index'],tw['toss_winner'])
plt.title("Toss Winner")
plt.xlabel("Teams")
plt.show()
```

# 2)Analysis on teams won after batting 1st

```python
#extracting records where team won after batting first
win_bat_first = ipl[ipl['win_by_runs']!=0]
win_bat_first
```

```python
plt.figure(figsize=(7,5))
plt.hist(win_bat_first['win_by_runs'])
plt.title("Winning by runs")
plt.xlabel("Runs")
plt.show()
```

```python
winner = pd.DataFrame(win_bat_first['winner'].value_counts()[0:3]).reset_index()
winner
```

```python
plt.figure(figsize=(7,5))
plt.bar(winner['index'],winner['winner'],color=['b','y','r'])
plt.title("Top 3 teams win after batting first")
plt.show()
```

# 3)Analysis on teams won after batting 2nd

```python
win_bat_second = ipl[ipl['win_by_wickets']!=0]
win_bat_second
```

```python
plt.figure(figsize=(9,5))
plt.hist(win_bat_second['win_by_wickets'],bins=30)
plt.title("Winning by wickets")
plt.xlabel("Wickets")
plt.show()
```

```python
#frequency of number of win
winner2 = pd.DataFrame(win_bat_second['winner'].value_counts()[0:3]).reset_index()
winner2
```

```python
plt.figure(figsize=(7,5))
plt.bar(winner2['index'],winner['winner'],color=['b','y','r'])
plt.title("Top 3 teams win after batting Second")
plt.show()
```
